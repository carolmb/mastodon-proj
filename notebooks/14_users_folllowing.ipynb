```python
import glob
import pandas as pd
```

```python
files = glob.glob('temp/users_following_kuni_*.tsv')
data = []
for file in files:
    temp = pd.read_csv(file, sep='\t')
    data.append(temp)

data = pd.concat(data)
data.columns = ['idx', 'source_acct', 'id', 'target_acct', 'locked', 'bot', 'discoverable', 'created_at', 'note', 'followers_count', 'following_count', 'statuses_count', 'last_status_at', 'fields']
print(data.head())
```

```python
data.to_csv('users_mstd_kuni_following.tsv', sep='\t', index=False)
```
