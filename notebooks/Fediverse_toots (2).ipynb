```python
import requests
import pandas as pd

date_limit = pd.Timestamp('2023-07-20', tz='utc')

def user_toots(instance, access_token):
    URL = f'https://{instance}/api/v1/timelines/public'
    user_toots = []

    if access_token:
        headers = {
            'Authorization': f'Bearer {access_token}'
        }
        is_end = False
        results = []
        max_id = None

        while not is_end:
            params = {
                'limit': 40,  # Increase the limit to fetch more toots per request (max 40).
                'max_id' : max_id
            }


            r = requests.get(URL, headers=headers, params=params)
            toots = r.json()

            if not isinstance(toots, list) or len(toots) == 0:
                break

            for t in toots:
                if 'created_at' in t:
                    timestamp = pd.Timestamp(t['created_at'])
                    if timestamp <= date_limit:
                        is_end = True
                        break

                    results.append(t)

            if is_end:
                break

            # Get the ID of the oldest toot in the current batch
            max_id = min(toot['id'] for toot in toots)

        df = pd.DataFrame(results)

        return df

    return None
```
